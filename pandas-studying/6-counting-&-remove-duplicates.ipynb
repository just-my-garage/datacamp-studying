# Counting values in a DataFrame

```python
import pandas as pd
california_housing_test = pd.read_csv('../datasets-from-colab/california_housing_test.csv')

sample_df = pd.DataFrame(california_housing_test)
sample_df
```

```python
sample_df['latitude'].value_counts()
```

```python
sample_df['latitude'].value_counts(sort=True)
```

```python
sample_df['latitude'].value_counts(normalize=True)
```

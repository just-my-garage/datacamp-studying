# Explicit Indexes
Why use them? It makes subsetting more easier.

Refer to <a href='./3-sorting-and-subsetting.ipynb'>Sorting and Subsetting</a> page, **last cell**, to get a better understanding 

```python
import pandas as pd

dataset = pd.read_excel('../datasets-from-colab/SampleData.xlsx', sheet_name=1)
dataset.head()
```

```python
"""
  EXAMPLE

    Using explicit index for bettwing viewing of `.isin()` method      
    (It was never tricky at the first place lol)
"""
column_as_index = dataset.set_index('Rep')
dataset[dataset['Rep'].isin(['Gill', 'Sorvino'])].head()
```

```python
"""
    SET/NEW INDEX

    <dataframe>.set_index(value[])
    <dataframe>.set_index('value')
"""
column_as_index = dataset.set_index(['Region', 'Rep'])
column_as_index.head()
```

```python
# REMOVE/RESET AN INDEX
reset_column_index = column_as_index.reset_index() # this works dumbo
reset_column_index.head()
```

```python
# DROP/DELETE INDEX
drop_column_index = column_as_index.reset_index(drop=True)
drop_column_index.head()
```

```python
"""
  SORT BY INDEX VALUES
  
  <dataframe>.sort_index(level?=[], ascending?=[])
"""
sort_column_index =  column_as_index.sort_index()
sort_column_index.head()
```

```python
sort_parameter_column_index =  column_as_index.sort_index(level=['Rep', 'Region'], ascending=[False, True])
sort_parameter_column_index.head()
```

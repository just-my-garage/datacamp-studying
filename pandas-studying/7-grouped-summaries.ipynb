# Grouped Summaries for Data

```python
import pandas as pd
california_housing_test = pd.read_csv('../datasets-from-colab/california_housing_test.csv')

sample_df = pd.DataFrame(california_housing_test)
sample_df
```

```python
# COLUMNS BECOME ROWS, AND ROWS BECOME COLUMNS
sample_df.groupby('population')['households'].median()
```

```python
# Get the mean of 'population; based on condition, with another filter 'households'
# COLUMNS BECOME ROWS, AND ROWS BECOME COLUMNS
sample_df[sample_df['population'] == 32]['households'].sum()
```

```python
# COLUMNS BECOME ROWS, AND ROWS BECOME COLUMNS
sample_df.groupby(['population', 'total_bedrooms'])[['households', 'total_bedrooms']].agg(['min', 'max', 'sum'])
```
